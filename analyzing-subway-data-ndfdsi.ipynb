              planejamento ou solução de problemas.

EXITS    = A contagem cumulativa de saídas associadas à catraca desde o último registro

### *Exercicio 1.2*

Escreva uma função que pegue a lista de nomes dos arquivos que você baixou no exercicio 1.1 e consolide-os em um único arquivo. Deve existir apenas uma linha de cabeçalho no arquivo de saida. 

Por exemplo, se o arquivo_1 tiver:
linha 1...
linha 2...

e o outro arquivo, arquivo_2 tiver:
linha 3...
linha 4...
linha 5...

Devemos combinar o arquivo_1 com arquivo_2 em um arquivo mestre conforme abaixo:

'C/A, UNIT, SCP, DATEn, TIMEn, DESCn, ENTRIESn, EXITSn'
linha 1...
linha 2...
linha 3...
linha 4...
linha 5...

**OBS:** Note que algumas colunas foram descartadas!

````python
def create_master_turnstile_file(input_files, output_file="merged_data.txt"):
    """
        It takes every files from input_files and merge its content in a single output_file.
        It'll ignore the first line of every input_file because they're labels.

        input_files: List. List of paths of each file you want to merge.
        output_file: String. Path to the merged file
    """
    
    with open(output_file, 'w') as master_file:
        master_file.write('C/A,UNIT,SCP,DATEn,TIMEn,DESCn,ENTRIESn,EXITSn\n')
        
        for input_file in input_files:
            with open(input_file, 'r', newline='') as file:
                for i, line in enumerate(file):
                    if i == 0:
                        continue
                        
                    data = line.strip().split(',')
                    to_export = []
                    to_export.extend(data[:3])
                    to_export.extend(data[6:])
                    
                    master_file.write(','.join(to_export) + '\n')
````

````python
output_file = "merged_data.txt"

exists = os.path.isfile(output_file)
if not exists:
    # Create a list of each input file path
    input_files = [folder + "/" + filename for filename in filenames]

    # Merge the files from the above list
    create_master_turnstile_file(input_files, output_file=output_file)
````

O dataframe deverá ficar conforme abaixo depois de filtrar apenas as linhas onde a coluna DESCn possua o valor REGULAR:

    0,A002,R051,02-00-00,05-01-11,00:00:00,REGULAR,3144312,1088151
    2,A002,R051,02-00-00,05-01-11,08:00:00,REGULAR,3144353,1088177


````python
def filter_by_regular(input_file):
    """
        It takes the file merged by `create_master_turnstile_file function`.
        Then creates a pandas Dataframe with only the samples which `DESCn` is `regular`.
        
        input_file: String. Merged file path.
    """
    
    turnstile_data = {}
    
    with open(input_file, 'r') as file:
        for i, line in enumerate(file):
            data = line.strip().split(',')
            
            if i == 0:
                columns = data
                
                for column in columns:
                    turnstile_data[column] = []
                
            elif data[5].lower() != "regular":
                continue
                
            else:
                for ii, column in enumerate(columns):
                    turnstile_data[column].append(data[ii])
            
    return pd.DataFrame(turnstile_data)
````

````python
def get_hourly_entries(df, inplace=False):
    """
        It takes a pandas subway Dataframe then creates a new label named `ENTRIESn_hourly`.
        This new label is the subtraction of the `ENTRIESn` column of the current row by the same column from the previous row.
        The first row gets a `ENTRIESn_hourly` equal to 1.
        
        df: pandas subway Dataframe
        inplace: Boolean. True: the new column is going to be attached to the original Dataframe. 
                          False: the new column is going to be attached to a copy of the original Dataframe.
    """
    
    if inplace is False:
        df = df.copy()
        
    df['dummy_var'] = df.shift(periods=1, fill_value="-1")['ENTRIESn']  
    df['ENTRIESn_hourly'] = df.apply(lambda x: int(x["ENTRIESn"]) - int(x["dummy_var"]) if x["dummy_var"] != "-1" else "1", 
                                     axis=1) 
    df.drop("dummy_var", axis=1, inplace=True)
    
    return df
````

### *Exercicio 2.3*

Calcule se estava nebuloso ou não (0 ou 1) e a temperatura máxima para fog (isto é, a temperatura máxima 
     para dias nebulosos).

````python
def max_temp_aggregate_by_fog(df, is_fog, to_report=False):
    """
        It returns the max value of `maxtempi` column for samples which contains or not fog.
        
        df: pandas Dataframe. It must have `fog` and `maxtempi` labels.
        is_fog: Boolean. True: It queries only samples with fog.
                         False: It queries only samples with no fog.
        to_report: Boolean. If true, it prints the max temperature in farenheit and celsius.
    """
    
    max_temp_F = np.max(df.query("fog == {}".format(float(is_fog)))['maxtempi'])
    
    if to_report == True:
        max_temp_C = round((max_temp_F - 32) * 5/9, 2)
        report = "Max Temperature: {} degrees Farenheit   |   {} degrees Celsius".format(max_temp_F, max_temp_C)
        print(report)
        
    return max_temp_F
````

````python
max_temp_fog = max_temp_aggregate_by_fog(df_weather, is_fog=True)
max_temp_no_fog = max_temp_aggregate_by_fog(df_weather, is_fog=False)

print("On foggy days, the maximum temperature was {} degrees Farenheit.".format(max_temp_fog))
print("On clear days, the maximum temperature was {} degrees Farenheit.".format(max_temp_no_fog))
````

### *Exercicio 2.4

Calcule agora a média de 'meantempi' nos dias que são sábado ou domingo (finais de semana):

````python
def avg_weekend_temperature(df):
    """
        It returns the average mean temperature on weekends.
        
        df: pandas Dataframe. It must have `DATEn` and `meantempi` labels. `DATEn` must be formated like: "year-month-day"
    """
    
    df['dummy_variable'] = df['DATEn'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").weekday())   
    mean_temp_weekends = np.mean(df.query("dummy_variable == 5 | dummy_variable == 6")['meantempi'])
    df.drop('dummy_variable', axis=1, inplace=True)
    return mean_temp_weekends
````

````python
mean_weekend_temp = avg_weekend_temperature(df_weather)

print("On weekends, the mean temperature was {} degres Farenheit.".format(round(mean_weekend_temp, 2)))
````

````python
def avg_min_temperature(df):
    """
        It returns the minimum temperature of rainy days.
        
        df: pandas Dataframe. It must have `rain` and `mintempi` labels.
    """
    
    avg_min_temp_rainy = np.mean(df.query("rain == 1.0 & mintempi > 55")['mintempi'])   
    return avg_min_temp_rainy
````

````python
def mapper():
    reader = csv.reader(open('turnstile_data_master_with_weather.csv'), delimiter=",")
    writer = csv.writer(open('mapper_result.txt', 'w', newline=''), delimiter="\t")
    
    for i, line in enumerate(reader):
        if len(line) != 22 or i == 0:
            continue

        unit = line[1]
        entries_hourly = str(int(float(line[6])))
        writer.writerow([unit, entries_hourly])


mapper()
````

### *Exercicio 3.2*

Agora crie o redutor. Dado o resultado do mapeador do exercicio anterior, o redutor deve imprimir(Não retornar) uma linha por UNIT, juntamente com o número total de ENTRIESn_hourly.Ao longo de maio (que é a duração dos nossos dados), separados por uma guia. Um exemplo de linha de saída do redutor pode ser assim: 'R001 \ t500625.0'

Você pode assumir que a entrada para o redutor está ordenada de tal forma que todas as linhas correspondentes a uma unidade particular são agrupados. No entanto a saida do redutor terá repetição pois existem lojas que aparecem em locais diferentes dos arquivos.

Exporte seu redutor em um arquivo chamado reducer_result.txt e envie esse arquivo juntamente com a sua submissão.

````python
def reducer():
    reader = csv.reader(open('mapper_result.txt'), delimiter="\t")
    writer = csv.writer(open('reducer_result.txt', 'w', newline=''), delimiter="\t")

    old_unit = None
    total = 0
    
    for line in reader:
        if len(line) != 2:
            continue
            
        this_unit, value = line
        
        if old_unit and old_unit != this_unit:
            writer.writerow([old_unit, total])
            total = 0
            
        old_unit = this_unit
        total += int(value)
        
    if old_unit != this_unit:
        writer.writerow([old_unit, total])


reducer()
````
